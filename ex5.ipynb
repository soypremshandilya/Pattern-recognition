```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread(r"C:\Users\DELL\Desktop\PATTERN RECOGNITION\Pattern-recognition\question.jpg", cv2.IMREAD_GRAYSCALE)

_, thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(1, 2, 2), plt.imshow(thresholded_image, cmap='gray'), plt.title('Thresholded Image')

plt.show()
```

```python
from skimage.measure import label


labeled_image = label(thresholded_image)

fig, ax = plt.subplots()
ax.imshow(labeled_image, cmap='nipy_spectral')
ax.set_title('Region-based Segmentation')
plt.show()
```

```python
edges = cv2.Canny(image, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title('Edge-based Segmentation')
plt.show()
```
